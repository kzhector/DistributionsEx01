```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data00.csv into a DataFrame
df = pd.read_csv('data00.csv')
#plot the data using matplotlib
#df.plot()
#test the data for stationarity
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['data'])

#print the p-value
print('p-value: {}'.format(result[1]))

#plot the autocorrelation function
from statsmodels.graphics.tsaplots import plot_acf
#plot_acf(df['data'])
#calculate the first difference
df['diff'] = df['data'] - df['data'].shift(1)
#plot the first difference
df['diff'].plot()

```

```python
#take first difference of the first difference
df['diff2'] = df['diff'] - df['diff'].shift(1)
#plot the second difference
df['diff2'].plot()

#test the second difference for stationarity
result = adfuller(df['diff2'].dropna())
#print the p-value
print('p-value: {}'.format(result[1]))

#average of the data difference
print(df['diff2'].mean())
#take moving average of the data difference
df['moving_avg'] = df['diff2'].rolling(10).mean()
#plot the moving average
df['moving_avg'].plot()
```
